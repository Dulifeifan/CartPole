```
import gym
import numpy as np
from keras.layers import Dense, Input
from keras.models import Model, Sequential
import keras
from keras.optimizers import RMSprop ,adam
from keras import backend as K
from keras.layers import Dense, Activation, Convolution2D, MaxPooling2D, Flatten
from keras.optimizers import Adam
import tensorflow as tf
import random
from collections import deque
from keras.models import Model
import matplotlib.pyplot as plt

env = gym.make('CartPole-v0')   
```

```
EPISODE = 300
np.random.seed(2)
MINIBATCH_SIZE=32
REPLAY_MEMORY_SIZE=2000
AGENT_HISTORY_LENGTH=4
TARGET_NETWORK_UPDATE_FREQUENCY=100
DISCOUNT_FACTOR=0.97
ACTION_REPEAT=4
UPDATE_FREQUENCY=4
LEARNING_RATE=0.001
GRADIENT_MOMENTUM=0.95
SQUARED_GRADIENT_MOMENTUM=0.95
MIN_SQUARED_GRADIENT=0.01
INITIAL_EXPLORATION=1
FINAL_EXPLORATION=0.01
FINAL_EXPLORATION_FRAME=1000000
REPLAY_START_SIZE=50000
NO_OP_MAX=30
```

```
class DQN:
  def __init__(self, feature_of_states,num_of_actions):
    self.feature_of_states=feature_of_states
    self.num_of_actions=num_of_actions
    self.lr=LEARNING_RATE
    self.gamma=DISCOUNT_FACTOR
    self.epsilon_max=INITIAL_EXPLORATION
    self.epsilon_min=FINAL_EXPLORATION
    self.replace_iter=50
    self.memory_size=REPLAY_MEMORY_SIZE
    self.batch_size=MINIBATCH_SIZE
    self.memory_iter = 0
    self.epsilon_now=0.1
    self.epsilon_change=0.001
    
  

    self.memory=np.zeros((REPLAY_MEMORY_SIZE, feature_of_states+1+1+feature_of_states)) #s,a,r,s'
    
    self.times_of_learn=0 #used to judge if to replace target
    self._build_evaluation_network()
    self._build_target_network()
  
  def _build_evaluation_network(self):
  
    model = Sequential()
    model.add(Dense(64, input_dim=feature_of_states, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(self.num_of_actions, activation='linear'))
    self.model_evaluation=model
    adam = Adam(lr=self.lr)
    self.model_evaluation.compile(loss=tf.losses.huber_loss, optimizer='adam', metrics=['mean_squared_error'])

  def _build_target_network(self):
    model = Sequential()
    model.add(Dense(64, input_dim=feature_of_states, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(self.num_of_actions, activation='linear'))
    self.model_target=model
    adam = Adam(lr=self.lr)
    self.model_target.compile(loss=tf.losses.huber_loss, optimizer='adam', metrics=['mean_squared_error'])
  
  def add_into_memory(self,state, action, reward,s_prime):
    #should deal with the situation that the memory is full
    self.memory[self.memory_iter % self.memory_size,:]=np.hstack((state,[action, reward],s_prime))
    self.memory_iter=self.memory_iter+1

    
  def get_next_action(self,state):
    state = np.array(state)
    state = state[np.newaxis, :]
    
    if(np.random.rand())>=self.epsilon_now:
      
      q_s_a=self.model_evaluation.predict(state)
      
      chosen_action=np.argmax(q_s_a[0])
    else:
      chosen_action=random.randrange(self.num_of_actions)
    return chosen_action
  
  def train(self):
    #judge if need to replace the weights
    if self.times_of_learn%self.replace_iter==0:
      parameter=self.model_evaluation.get_weights()
      self.model_target.set_weights(parameter)
    #randomly choose batch_size memory
    random_chosen_memory = self.memory[np.random.choice(min(self.memory_size,self.memory_iter),size=self.batch_size)]
    s=random_chosen_memory[:, :4]
    s_prime=random_chosen_memory[:, -4:]
    a=random_chosen_memory[:, 4].astype(int)
    r=random_chosen_memory[:, 5]
    q=self.model_evaluation.predict(s)
    q_prime=self.model_target.predict(s_prime)
    target_array=q.copy()
    index = np.arange(self.batch_size, dtype=np.int32)
    target_array[index,a]=r+self.gamma*np.max(q_prime,axis=1)
    self.model_evaluation.fit(s,target_array,epochs=1,verbose=0)
    if self.epsilon_now>self.epsilon_min:
      self.epsilon_now=self.epsilon_now-self.epsilon_change
    else:
      self.epsilon_now=self.epsilon_min
    self.times_of_learn=self.times_of_learn+1

```

```
if __name__ == "__main__":
  feature_of_states=env.observation_space.shape[0]
  num_of_actions=env.action_space.n
  model=DQN(feature_of_states,num_of_actions)
  reward_record=[]
  i=0 #control when to learn
  for episode in range(EPISODE):
    total_reward=0
    state = env.reset() #initial env
    #print(state)
  
    #print(state)
    for a in range (5000):
    #while (1):
      action=model.get_next_action(state)
      s_prime, reward, done, _ =env.step(action)
      total_reward=total_reward+reward

      location_x_axis, _, theta_of_cart_pole, _ = s_prime
      location_reward = (env.x_threshold - abs(location_x_axis))/env.x_threshold - 0.8
      theta_reward = (env.theta_threshold_radians - abs(theta_of_cart_pole))/env.theta_threshold_radians - 0.5
      reward = location_reward + theta_reward   

    
      # if done:
      #   reward=-20
      # else:
      #   reward =reward
      
      model.add_into_memory(state, action, reward,s_prime)
      if (MINIBATCH_SIZE<len(model.memory)):
        if i>500:
          model.train()
      
      
      if done:
        print(episode,total_reward)
        reward_record.append(total_reward)
        break
      state=s_prime
      i=i+1

  print("Finished!")
  reward_array=np.array(reward_record)
  plt.plot(np.convolve(reward_array, np.ones((100,)) / 100, mode='valid')) 
  plt.ylabel('Moving Average of Reward')
  plt.xlabel('Training Episode')
  plt.show()
  print(reward_array)
  plt.plot(reward_record) 
  plt.ylabel('Reward')
  plt.xlabel('Training Episode')
  plt.show()
```
